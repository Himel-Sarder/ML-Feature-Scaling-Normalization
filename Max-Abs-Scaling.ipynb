```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("placement.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
plt.scatter(df['cgpa'],df['package'], c='r')
plt.xlabel('CGPA')
plt.ylabel('PACKAGE')
```

```python
# Custom MaxAbs Scaling Function
def max_abs_scale(X):
    max_abs = np.max(np.abs(X), axis=0)
    return X / max_abs

# Apply to the example data
scaled_data = max_abs_scale(df)
scaled_df = pd.DataFrame(scaled_data, columns=df.columns)
scaled_df
```

```python
plt.scatter(scaled_df['cgpa'],scaled_df['package'], c='r')
plt.xlabel('CGPA')
plt.ylabel('PACKAGE')
```

## Using sklearn for MaxAbs Scaling

```python
from sklearn.preprocessing import MaxAbsScaler
```

```python
# Initialize the MaxAbsScaler
scaler = MaxAbsScaler()
```

```python
# Fit and transform the data
scaled_data = scaler.fit_transform(df)
```

```python
# Convert back to DataFrame
scaled_df = pd.DataFrame(scaled_data, columns=df.columns)
```

```python
scaled_df
```

```python
plt.scatter(scaled_df['cgpa'],scaled_df['package'], c='r')
plt.xlabel('CGPA')
plt.ylabel('PACKAGE')
plt.title("After Scaling")
```
