# ***Mean-Max Scaling***

Mean-Max Scaling is a data normalization technique where each feature is scaled using the mean and the maximum value of the feature. The formula for Mean-Max Scaling is:

## Import necessary libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Read CSV File

```python
df = pd.read_csv("placement.csv")
```

```python
df.head()
```

```python
df.sample(10)
```

```python
df.info()
```

## Plot data

```python
plt.scatter(df['cgpa'],df['package'], c='r')
plt.xlabel('CGPA')
plt.ylabel('PACKAGE')
```

## Function for perform mean max scaling

```python
def mean_max_scaling(column):
    mean = column.mean()
    max_value = column.max()
    return (column - mean) / max_value

# Apply Mean-Max Scaling
scaled_data = df.apply(mean_max_scaling)
scaled_data
```

## Plot modified data

```python
plt.scatter(scaled_data['cgpa'],scaled_data['package'], c='r')
plt.xlabel('CGPA')
plt.ylabel('PACKAGE')
```
