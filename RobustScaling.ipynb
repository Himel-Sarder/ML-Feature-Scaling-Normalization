```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("placement.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
plt.scatter(df['cgpa'],df['package'], c='g')
plt.xlabel('CGPA')
plt.ylabel('PACKAGE')
```

```python
# Custom Robust Scaling Function
def robust_scale(X):
    Q1 = np.percentile(X, 25, axis=0)  # 25th percentile
    Q3 = np.percentile(X, 75, axis=0)  # 75th percentile
    IQR = Q3 - Q1
    median = np.median(X, axis=0)
    return (X - median) / IQR

# Apply to the example data
scaled_data = robust_scale(df)
scaled_df = pd.DataFrame(scaled_data, columns=df.columns)
scaled_df
```

```python
plt.scatter(scaled_df['cgpa'],scaled_df['package'], c='b')
plt.xlabel('CGPA')
plt.ylabel('PACKAGE')
```

## Using sklearn for Robust Scaling

```python
from sklearn.preprocessing import RobustScaler
```

```python
# Initialize the RobustScaler
scaler = RobustScaler()
```

```python
# Fit and transform the data
scaled_data = scaler.fit_transform(df)
```

```python
# Convert back to DataFrame
scaled_df = pd.DataFrame(scaled_data, columns=df.columns)
scaled_df
```

```python
plt.scatter(scaled_df['cgpa'],scaled_df['package'], c='y')
plt.xlabel('CGPA')
plt.ylabel('PACKAGE')
```
